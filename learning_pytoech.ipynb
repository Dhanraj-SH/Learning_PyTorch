# Tensors

Tensors are a specialized data structure that are very similar to arrays and matrices. In PyTorch, we use tensors to encode the inputs and outputs af a model, as well as the models parameters.

Tensors are similar to NumPy’s ndarrays, except that tensors can run on GPUs or other specialized hardware to accelerate computing.

```python
import torch
import numpy as np
```

#### Tensor Initialization 

Tensors can be initialized in various ways

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
```

```python
x_data
```

```python
#### From a NumPy array - Tensors can be created from NumPy arrays and vice versa 
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

```python
x_np
```

```python
#From another tensor: The new tensor retains the properties(shape, datatype) of the argument tensor, useless explicitily overridden.
x_ones = torch.ones_like(x_data)
x_ones
```

```python
x_rand = torch.rand_like(x_data, dtype = torch.float)
x_rand
```

```python
#With random and constant values - 'shape' is a tuple dimensions. In the functions below, it determines the dimensionality of the output tensor
shape = (2, 3,)

rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(rand_tensor)
print(ones_tensor)
print(zeros_tensor)
```

#### Tensor Operations

Over 100 tensor operations, including transpoing, indexing, slicing, mathematical operations, linear algebra, random sampling and more are somprehensively described

Each of them can be run on the GPU (at typically higher speeds than on a CPU). If you're usin gColab, allocate a GPU by going to

```python
# We move our tensor to the GPU if available
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
    tensor.device
```

```python
tensor = torch.ones(4, 4)
tensor[:,1] = 0
tensor 
```

Joining tensors - You can use `torch.cat` to concatenate a sequence of tensors along a given dimension. And also there is `torch.stack`, another tensor joining op that is subtly different from `torch.cat`

```python
t1 = torch.cat([tensor, tensor, tensor], dim = 1)
t1
```

#### Multiplying tensors 
This computes the element-wise product

```python
tensor.mul(tensor)
```

```python
#alternative element-wise syntax:
tensor * tensor
```

```python
# This computes the matrix multiplication between two tensors
tensor.matmul(tensor.T)
```

```python
#Alternative method
tensor@tensor.T
```

#### In-place operations

Operations that have a `_` suffix are in-place. For example: `x.copy_(y)`, `x.t_()` will change `x`.

```python
tensor
```

```python
tensor.add_(5)
tensor
```

__NOTE__ :

In-place operations save some memory, but can be problematic when computing derivatives because of an immediate loss of history.
Hence there use is discouraged.
